```python
import numpy as np
import matplotlib.pyplot as plt
# Result collection\final_25epoch_collection\1 frame per second, Resnet 152, dropout 0.1, epoch 25
```

```python
train_loss = np.load('CRNN_epoch_training_loss.npy')
train_accuracy = np.load('CRNN_epoch_training_accuracy.npy')
test_loss = np.load('CRNN_epoch_test_loss.npy')
test_accuracy = np.load('CRNN_epoch_test_accuracy.npy')
```

```python
epochs = 25
fig = plt.figure(figsize=(20, 8))
plt.subplot(121)
plt.plot(np.arange(1, epochs + 1), train_loss[:, -1], 'r')  # train loss (on epoch end)
plt.plot(np.arange(1, epochs + 1), test_loss, 'g')         #  test loss (on epoch end)
plt.title("model loss")
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend(['train', 'test'], loc="upper left")
```

```python
fig = plt.figure(figsize=(20, 8))
plt.subplot(122)
plt.plot(np.arange(1, epochs + 1), train_accuracy[:, -1], 'r')  # train accuracy (on epoch end)
plt.plot(np.arange(1, epochs + 1), test_accuracy, 'g')         #  test accuracy (on epoch end)
plt.title("training scores")
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend(['train', 'test'], loc="upper left")
title = "./fig_UCF101_ResNetCRNN.png"
plt.savefig(title, dpi=600)
# plt.close(fig)
plt.show()
```
